# Grammar Error Detection — Student-1 / Lesson-2

Runs LanguageTool on each sentence and classifies errors into the four **Proficiency Dimensions**:
- **A** — Sentence Architecture
- **B** — Tense & Aspect Mastery
- **C** — Nominal Precision
- **D** — Modal & Functional Range

```python
import sys, pathlib

# Locate repo root (walk up from notebooks/)
here = pathlib.Path().resolve()
root = here
for _ in range(4):
    if (root / 'grammar_errors').is_dir():
        break
    root = root.parent
if str(root) not in sys.path:
    sys.path.insert(0, str(root))

from grammar_errors import ErrorChecker

TARGET = root / 'processed_data' / 'sentences' / 'Student-1' / 'lesson-2' / 'validated-sentence-separation.txt'
n_lines = sum(1 for l in TARGET.read_text(encoding='utf-8').splitlines() if l.strip())
print(f'Target: {TARGET}')
print(f'Sentences: {n_lines}')
```

```python
checker = ErrorChecker()
errors = checker.check_file(TARGET)
checker.close()
print(f'Grammar errors detected: {len(errors)}')
```

```python
from IPython.display import display, HTML
from collections import Counter

DIM_COLORS = {
    'A': {'bg': '#DBEAFE', 'border': '#3B82F6', 'badge': '#1D4ED8', 'label': 'Sentence Architecture'},
    'B': {'bg': '#FEF3C7', 'border': '#F59E0B', 'badge': '#92400E', 'label': 'Tense & Aspect Mastery'},
    'C': {'bg': '#D1FAE5', 'border': '#10B981', 'badge': '#065F46', 'label': 'Nominal Precision'},
    'D': {'bg': '#EDE9FE', 'border': '#8B5CF6', 'badge': '#4C1D95', 'label': 'Modal & Functional Range'},
}


def highlight_error(sentence, offset, length, dim):
    col = DIM_COLORS[dim]['border']
    before = sentence[:offset]
    match = sentence[offset:offset + length]
    after = sentence[offset + length:]
    style = 'background:' + col + '33;border-bottom:2px solid ' + col + ';padding:1px 0;'
    return before + '<mark style="' + style + '"><b>' + match + '</b></mark>' + after


def render_errors(errors):
    if not errors:
        display(HTML('<p style="color:#6B7280;font-family:sans-serif;">No grammar errors detected.</p>'))
        return

    rows = []
    for e in errors:
        c = DIM_COLORS[e['dimension_code']]
        badge_style = ('background:' + c['badge'] + ';color:white;padding:2px 8px;'
                       'border-radius:10px;font-size:11px;font-weight:700;')
        dim_badge = '<span style="' + badge_style + '" title="' + c['label'] + '">' + e['dimension_code'] + '</span>'

        highlighted = highlight_error(e['sentence'], e['offset'], e['error_length'], e['dimension_code'])
        replacements = ', '.join('<i>' + r + '</i>' for r in e['replacements'][:3]) or '<span style="color:#9CA3AF">—</span>'

        row_style = ('background:' + c['bg'] + ';border-left:4px solid ' + c['border'] + ';'
                     'vertical-align:top;border-bottom:1px solid #E5E7EB;')
        td = 'style="padding:8px 12px;"'
        rows.append(
            '<tr style="' + row_style + '">'
            + '<td ' + td + ' style="padding:8px 12px;color:#6B7280;font-size:12px;">' + str(e['sentence_index']) + '</td>'
            + '<td ' + td + ' style="padding:8px 12px;font-family:Georgia,serif;line-height:1.7;">' + highlighted + '</td>'
            + '<td ' + td + ' style="padding:8px 12px;white-space:nowrap;">' + dim_badge + '</td>'
            + '<td ' + td + ' style="padding:8px 12px;font-size:12px;font-weight:600;">' + e['grammar_category'] + '</td>'
            + '<td ' + td + ' style="padding:8px 12px;font-size:12px;color:#374151;">' + e['message'][:90] + '</td>'
            + '<td ' + td + ' style="padding:8px 12px;font-size:12px;color:#6B7280;">' + replacements + '</td>'
            + '</tr>'
        )

    th = 'style="padding:8px 12px;text-align:left;"'
    header = (
        '<tr style="background:#F3F4F6;font-size:11px;font-weight:700;text-transform:uppercase;color:#374151;">'
        + '<th ' + th + '>#</th>'
        + '<th ' + th + '>Sentence  (error highlighted)</th>'
        + '<th ' + th + '>Dim</th>'
        + '<th ' + th + '>Grammar category</th>'
        + '<th ' + th + '>Explanation</th>'
        + '<th ' + th + '>Suggestions</th>'
        + '</tr>'
    )
    html = (
        '<table style="border-collapse:collapse;width:100%;font-size:13px;font-family:sans-serif;">'
        + header + ''.join(rows) + '</table>'
    )
    display(HTML(html))


render_errors(errors)
```

```python
def render_summary(errors):
    if not errors:
        display(HTML('<p style="font-family:sans-serif;">No errors.</p>'))
        return

    total = len(errors)
    dim_counts = Counter(e['dimension_code'] for e in errors)
    cat_counts = Counter((e['dimension_code'], e['grammar_category']) for e in errors)

    # Dimension bar chart
    bars = []
    for dim in ['A', 'B', 'C', 'D']:
        n = dim_counts.get(dim, 0)
        pct = n / total * 100 if total else 0
        c = DIM_COLORS[dim]
        bar_inner = '<div style="width:' + str(round(pct, 1)) + '%;background:' + c['border'] + ';height:20px;border-radius:3px;"></div>'
        bars.append(
            '<div style="display:flex;align-items:center;gap:12px;margin-bottom:8px;">'
            + '<span style="width:22px;font-weight:800;font-size:14px;color:' + c['badge'] + '">' + dim + '</span>'
            + '<span style="font-size:12px;width:210px;color:#374151;">' + c['label'] + '</span>'
            + '<div style="flex:1;background:#E5E7EB;border-radius:3px;height:20px;">' + bar_inner + '</div>'
            + '<span style="width:36px;text-align:right;font-weight:700;font-size:13px;">' + str(n) + '</span>'
            + '</div>'
        )

    # Category breakdown table
    cat_rows = []
    for (dim, cat), count in sorted(cat_counts.items(), key=lambda x: (-x[1], x[0][0])):
        c = DIM_COLORS[dim]
        badge_style = 'background:' + c['badge'] + ';color:white;padding:1px 7px;border-radius:8px;font-size:11px;font-weight:700;'
        cat_rows.append(
            '<tr style="border-bottom:1px solid #F3F4F6;">'
            + '<td style="padding:5px 10px;"><span style="' + badge_style + '">' + dim + '</span></td>'
            + '<td style="padding:5px 10px;font-size:13px;">' + cat + '</td>'
            + '<td style="padding:5px 10px;font-weight:700;text-align:right;font-size:13px;">' + str(count) + '</td>'
            + '</tr>'
        )

    html = (
        '<div style="font-family:sans-serif;max-width:700px;">'
        + '<h3 style="margin:0 0 16px;">Summary &mdash; ' + str(total) + ' errors detected</h3>'
        + '<div style="margin-bottom:24px;">' + ''.join(bars) + '</div>'
        + '<h4 style="margin:0 0 8px;">Breakdown by grammar category</h4>'
        + '<table style="border-collapse:collapse;font-size:13px;min-width:350px;">'
        + '<tr style="background:#F3F4F6;font-size:11px;font-weight:700;text-transform:uppercase;">'
        + '<th style="padding:6px 10px;text-align:left;">Dim</th>'
        + '<th style="padding:6px 10px;text-align:left;">Grammar category</th>'
        + '<th style="padding:6px 10px;text-align:right;">Count</th>'
        + '</tr>'
        + ''.join(cat_rows)
        + '</table></div>'
    )
    display(HTML(html))


render_summary(errors)
```
